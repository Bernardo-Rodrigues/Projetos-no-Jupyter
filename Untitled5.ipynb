## Análise de dados(Expectativa de vida)

O desafio é fazer uma análise sobre a expectativa de vida da papolação em países de todo o mundo entre os anos de 1802 e 2016

```python
import pandas as pd

tabela = pd.read_csv("Life expectancy.csv")
display(tabela)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,6))
sns.lineplot(x = tabela["Year"], y = tabela["Life expectancy"])
plt.show()
plt.figure(figsize=(20,6))
sns.barplot(x = tabela["Entity"], y = tabela["Life expectancy"])
plt.show()
```

## Conclusões

É possível notar que durante o avanço dos anos, principalmente depois de 1850, a expectativa de vida foi aumentando, com exceção em apenas 2 periodos que são exatamente os periodos de guerra mundial, a primeira irregularidade no grafico é devido à primeira guerra(1914-1918) e a segunda irregularidade é devido à segunda guerra(1939-1945). 

É possível ver também pelo gráfico dos países que os localizados na Europa e na América do Norte possuem uma média de expectativa de vida maior do que os da Ásia e resto da América.
